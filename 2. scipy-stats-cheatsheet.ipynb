## 2. scipy.stats Cheatsheet

```python
from scipy.stats import uniform
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. uniform distribution

dist1 = uniform(loc = 10, scale = 3)
# loc is the lower bound of the distribution
# scale is the width of the distribution

# create sample data
x = np.linspace(8, 15, 100)
y = dist1.pdf(x)


plt.plot(x, y)
plt.title('Uniform distribution')
plt.show()
```

```python
# 2. normal distribution

dist2 = norm(loc = 0, scale = 1)
# loc is the mean of the distribution
# scale is the standard deviation of the distribution

# create sample data
x = np.linspace(-5, 5, 100)
y = dist2.pdf(x)

plt.plot(x, y)
plt.title('Normal distribution')
plt.show()
```

```python
# 3. Describe the distribution
dist1_mean = dist1.mean()
dist1_var = dist1.var()
dist1_std = dist1.std()
dist1_mean, dist1_var, dist1_std
```

```python
# 4. Calculate the probability of a value (uniform distribution)
p_9 = dist1.pdf(9)
p_11 = dist1.pdf(11)
p_11_or_12 = dist1.pdf(11) + dist1.pdf(12)
p_9, p_11_or_12
```

```python
# 5. Calculate the probability of a value (normal distribution)
p_0 = dist2.pdf(0)
p_1 = dist2.pdf(1)
p_0_or_1 = dist2.pdf(0) + dist2.pdf(1)
p_0, p_1, p_0_or_1
```

```python
# 6. Calculate the cumulative probability (uniform distribution)
p_9 = dist1.cdf(9)
p_11 = dist1.cdf(11)
p_11_or_12 = dist1.cdf(12)
p_9, p_11_or_12

# same as the pdf in this case, because the distribution is uniform
```

```python
# 7. Calculate the cumulative probability (normal distribution)
p_0 = dist2.cdf(0)
p_1 = dist2.cdf(1)
p_0_or_1 = dist2.cdf(1)
p_0, p_1, p_0_or_1

# Different in this case, because the distribution is normal, and it is area under the curve to the left
```
