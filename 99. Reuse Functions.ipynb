# Reusable function for histogram

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_gaussian_distribution(data, label='Data', bins=30, color='lightblue', loc=0, scale=1, pdf_color='black', legend=False):
    """
    Plots a histogram of the data and overlays a Gaussian distribution.

    Parameters:
    - data: array-like, the data to plot
    - label: str, the label for the histogram (default is 'Data')
    - bins: int, the number of bins for the histogram (default is 30)
    - color: str, the color of the histogram (default is lightblue)
    - loc: float, the mean of the Gaussian distribution
    - scale: float, the standard deviation of the Gaussian distribution
    - pdf_color: str, the color of the Gaussian PDF line (default is black)
    """
    # Plot the histogram of the data
    counts, bin_edges, _ = plt.hist(data, bins=bins, color=color, label=label, alpha=0.6, edgecolor='black')

    # Calculate the area under the histogram
    bin_width = bin_edges[1] - bin_edges[0]
    area = sum(counts) * bin_width

    # Plot the Gaussian distribution using the PDF function, scaled by the area
    x = np.linspace(min(data), max(data), 1000)
    pdf = norm.pdf(x, loc=loc, scale=scale) * area
    plt.plot(x, pdf, color=pdf_color, lw=2, label='Gaussian PDF')

    # Add labels and legend
    if not label:
        label = 'Data'
    plt.xlabel(label)
    plt.ylabel('Frequency')
    if legend:
        plt.legend()

    # Show the plot
    plt.show()

# Example usage
data = np.random.normal(loc=0, scale=1, size=1000)
plot_gaussian_distribution(data, loc=data.mean(), scale=data.std())
```
