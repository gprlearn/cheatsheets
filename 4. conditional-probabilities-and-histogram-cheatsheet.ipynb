Using the `titanic` data, conditional probabilities can be calculated by subsetting the data to the condition of interest and comparing the outcomes within this criteria.  For example, what is the probability, given that someone is in first class, that they are over the age of 40?

```python
first_class = #how many people in first class

```

# Reusable function

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_gaussian_distribution(data, label='Data', bins=30, color='lightblue', loc=0, scale=1, pdf_color='black'):
    """
    Plots a histogram of the data and overlays a Gaussian distribution.

    Parameters:
    - data: array-like, the data to plot
    - label: str, the label for the histogram (default is 'Data')
    - bins: int, the number of bins for the histogram (default is 30)
    - color: str, the color of the histogram (default is lightblue)
    - loc: float, the mean of the Gaussian distribution
    - scale: float, the standard deviation of the Gaussian distribution
    - pdf_color: str, the color of the Gaussian PDF line (default is black)
    """
    # Plot the histogram of the data
    sns.histplot(data, kde=False, stat='density', bins=bins, color=color, label=label)

    # Plot the Gaussian distribution using the PDF function
    x = np.linspace(min(data), max(data), 1000)
    pdf = norm.pdf(x, loc=loc, scale=scale)
    plt.plot(x, pdf, color=pdf_color, lw=2, label='Gaussian PDF')


    # Add labels and legend
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()

    # Show the plot
    plt.show()

# Example usage
data = np.random.normal(loc=0, scale=1, size=1000)
plot_gaussian_distribution(data)

````

````python
plot_gaussian_distribution(gap_df['lifeExp'], bins=30, color='lightblue', pdf_color='black', label='Life Expectancy', loc=gap_df['lifeExp'].mean(), scale=gap_df['lifeExp'].std())
````
