## 3. corr cheatsheet

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

penguin_df = pd.read_csv('data/penguins.csv')
penguin_df.info()
```

```python
# Check the columns of the dataframe
print(penguin_df.columns)

# Select the desired columns if they exist
if 'bill_length_mm' in penguin_df.columns and 'bill_depth_mm' in penguin_df.columns:
	penguin_df = penguin_df[['bill_length_mm', 'bill_depth_mm']]
else:
	print("The required columns are not present in the dataframe.")
```

```python
# penguins_cov = penguin_df.cov()
# penguins_corr = penguin_df.corr()
sns.heatmap(penguin_df.corr(), annot = True)
```
